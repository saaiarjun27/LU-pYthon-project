```python
import random
def playAgain():
    pg=input("\nDo you wanna play again? (y/n) :")
    if pg=='y' or pg=='n':
        return pg
    else:
        return playAgain()
def guess():
    again='y'
    while again=='y':
        score=0
        computerNum=random.randint(0,100)
        userNum=int(input("Enter your guess:"))
        diff=computerNum-userNum
        if diff==0:
            print("Computer Number: ",computerNum)
            print("User Guess: ",userNum)
            print("\nCongrats you have guessed the exact number!!!")
            score=score+50
            print("\nYour Score:",score)
        elif diff>= -10 and diff<=10:
            print("Computer Number: ",computerNum)
            print("User Guess: ",userNum)
            print("\nYou are close to the number!!!")
            score=score+10
            print("\nYour Score:",score)
        elif diff>= -5 and diff<=5:
            print("Computer Number: ",computerNum)
            print("User Guess: ",userNum)
            print("\nYou are very close to the number!!!")
            score=score+25
            print("\nYour Score:",score)
        else:
            print("Computer Number: ",computerNum)
            print("User Guess: ",userNum)
            print("\nWrong Guess, Better luck next time!!!")
            score=score+0
            print("\nYour Score:",score)
        again=playAgain()
    else:
        print("\nFinal score:", score)
guess()
            
```
